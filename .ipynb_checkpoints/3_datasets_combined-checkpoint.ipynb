## Data Preprocessing & Augmentation

- This cell **loads and preprocesses the dataset** for glaucoma detection. It reads images from the dataset, resizes them, and assigns labels (**normal** or **glaucoma**). The data is then **shuffled for randomness**, ensuring better generalization during training.  

- A **data augmentation pipeline** is applied, including **flips, rotations, contrast adjustments, and translations**, helping the model become more robust to variations in input images.  

- Additionally, **CLAHE (Contrast Limited Adaptive Histogram Equalization)** is used to enhance image quality by improving contrast while preventing over-amplification of noise. It applies histogram equalization to small regions of an image, making subtle details more visible without distorting important features.  

- Finally, the processed images are **converted into a structured format** suitable for deep learning model training. 


```python
import os
import cv2
import numpy as np
import random
from tqdm import tqdm
from tensorflow.keras.layers import RandomFlip, RandomRotation, RandomContrast, RandomTranslation
from tensorflow.keras.models import Sequential
from sklearn.utils import shuffle

# Set seed for reproducibility
random.seed(1)
np.random.seed(1)

# Define dataset path and categories
DATADIR = r"C:\Users\Nishitha T\OneDrive\Mini Project\Mini project 2\Final Project\mixed_3_datasets\Training"
CATEGORIES = ["normal", "glaucoma"]

IMG_SIZE = 256
training_data = []

# Function to create training dataset
def create_training_data():
    for category in CATEGORIES:
        path = os.path.join(DATADIR, category)
        class_num = CATEGORIES.index(category)

        for img in tqdm(os.listdir(path)):
            try:
                img_array = cv2.imread(os.path.join(path, img), cv2.IMREAD_COLOR)  # Read image
                if img_array is None:
                    continue  # Skip if image is not read
                new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))  # Resize image
                training_data.append([new_array, class_num])
            except Exception as e:
                print(f"Error loading {img}: {e}")

# Run dataset creation
create_training_data()

# Shuffle the dataset
training_data = shuffle(training_data, random_state=1)

# Data Augmentation Pipeline
data_augmentation = Sequential([
    RandomFlip(mode="horizontal_and_vertical", seed=1),
    RandomRotation(factor=0.1, seed=1),  # Â±10 degrees
    RandomContrast(factor=0.1, seed=1),
    RandomTranslation(height_factor=0.15, width_factor=0.15, fill_mode='wrap', seed=1)
])
'''
# Function for Histogram Equalization
def he(img):
    ycrcb = cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb)  # Convert to YCrCb
    y, cr, cb = cv2.split(ycrcb)  # Split channels
    y_eq = cv2.equalizeHist(y)  # Apply histogram equalization on Y channel
    merged = cv2.merge((y_eq, cr, cb))  # Merge channels back
    return cv2.cvtColor(merged, cv2.COLOR_YCrCb2BGR)  # Convert back to BGR
'''
# Function for CLAHE (Contrast Limited Adaptive Histogram Equalization)
def clahe(img):
    lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)  # Convert to LAB
    l, a, b = cv2.split(lab)  # Split channels
    clahe = cv2.createCLAHE(clipLimit=3.5, tileGridSize=(8, 8))  # Define CLAHE
    l_clahe = clahe.apply(l)  # Apply CLAHE to L channel
    merged = cv2.merge((l_clahe, a, b))  # Merge LAB channels back
    return cv2.cvtColor(merged, cv2.COLOR_LAB2BGR)  # Convert back to BGR

# Apply Preprocessing
X = []
y = []

for features, label in training_data:
    features = clahe(features)  # Apply CLAHE
    X.append(features)
    y.append(label)

# Convert to NumPy arrays
X = np.array(X).reshape(-1, IMG_SIZE, IMG_SIZE, 3)
y = np.array(y)

print(f"Dataset loaded: {X.shape}, Labels: {y.shape}")
```

## Model Architecture: ResNet50 for Glaucoma Detection  

ResNet (Residual Network) is ideal for glaucoma detection due to its ability to capture deep and complex patterns in fundus images. Glaucoma affects the optic nerve, requiring precise identification of subtle changes in the optic disc and cup ratio, which ResNet50 effectively learns through its deep layers.

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.layers import RandomRotation, RandomTranslation

IMG_SIZE = 256
NUM_CLASSES = 2  # Two categories: 'normal' and 'glaucoma'

# Load ResNet50 as base model (pre-trained on ImageNet)
base_model = ResNet50(input_shape=(IMG_SIZE, IMG_SIZE, 3), include_top=False, weights='imagenet')

# Freeze all layers initially
base_model.trainable = False

# Unfreeze last few layers for fine-tuning
for layer in base_model.layers[-10:]:  # Fine-tune last 10 layers
    layer.trainable = True

# Define the model
model = Sequential([
    # Data augmentation layers (only applied during training) - Removed RandomFlip
    RandomRotation(0.1),  # Â±10 degrees
    RandomTranslation(height_factor=0.15, width_factor=0.15, fill_mode='wrap'),

    # Pre-trained ResNet50 as feature extractor
    base_model,

    # Feature extraction layer
    GlobalAveragePooling2D(),

    # Fully connected layers
    Dropout(0.5),  # Dropout before Dense to reduce overfitting
    Dense(256, activation='relu'),
    Dense(NUM_CLASSES, activation='softmax')  # Output layer
])

# Compile the model
model.compile(optimizer=Adam(learning_rate=1e-4),  # Learning rate written in scientific notation
              loss='sparse_categorical_crossentropy',  # Use if labels are integers (0,1)
              metrics=['accuracy'])

# Model summary
model.summary()
```

## Model Training

```python
from tensorflow.keras.callbacks import EarlyStopping

# training the model
batch_size = 32
epochs = 40
early_stop = EarlyStopping(monitor="val_loss", patience=3, restore_best_weights=True)
#history = model.fit(X, y, batch_size=batch_size, epochs=epochs, validation_split=0.15,callbacks=[early_stop])  # Stop early if no improvement
history = model.fit(X, y, batch_size=batch_size, epochs=10, validation_split=0.15)
'''
history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=40,  # Start with a high number
    callbacks=[early_stop]  # Stop early if no improvement
)
'''
```

## Model Evaluation

```python
train_acc = history.history['accuracy'][-1]
val_acc = history.history['val_accuracy'][-1]

print(f"Final Training Accuracy: {train_acc:.4f}")
print(f"Final Validation Accuracy: {val_acc:.4f}")
```

```python
# Define test dataset path
TEST_DIR = r"C:\Users\Nishitha T\OneDrive\Mini Project\Mini project 2\Final Project\mixed_3_datasets\Testing"

test_data = []
IMG_SIZE = 256
CATEGORIES = ["normal", "glaucoma"]

# Load and preprocess test data
for category in CATEGORIES:
    path = os.path.join(TEST_DIR, category)
    class_num = CATEGORIES.index(category)

    for img in os.listdir(path):
        try:
            img_array = cv2.imread(os.path.join(path, img), cv2.IMREAD_COLOR)
            if img_array is None:
                continue
            img_resized = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))  # Resize image
            img_preprocessed = clahe(img_resized)  # Apply same preprocessing as training data
            test_data.append([img_preprocessed, class_num])
        except Exception as e:
            print(f"Error loading {img}: {e}")

# Shuffle and convert test data to arrays
test_data = shuffle(test_data, random_state=1)
X_test = np.array([i[0] for i in test_data]).reshape(-1, IMG_SIZE, IMG_SIZE, 3)
y_test = np.array([i[1] for i in test_data])
```

```python
from tensorflow.keras.models import load_model

model = load_model("resnet50_model1.h5")
```

```python
test_loss, test_acc = model.evaluate(X_test, y_test, verbose=1)
print(f"Test Accuracy: {test_acc:.4f}")
print(f"Test Loss: {test_loss:.4f}")
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)  # Convert probabilities to class labels

print(classification_report(y_test, y_pred_classes, target_names=CATEGORIES))

# Plot confusion matrix
cm = confusion_matrix(y_test, y_pred_classes)
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, cmap="Blues", fmt="d", xticklabels=CATEGORIES, yticklabels=CATEGORIES)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```
